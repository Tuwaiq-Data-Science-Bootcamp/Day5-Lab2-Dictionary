```python
PB = {
    '1111111111': "Amal",
    '2222222222': "Mohammed",
    '3333333333': "Khadijah",
    '4444444444': "Abdullah", 
    '5555555555': "Rawan",  
    '6666666666': "Faisal", 
    '7777777777': "Layla"
}

def SN(name: str):
    count = 0 
    for num,names in PB.items():
        if names.lower() == name.lower():
            print("The number is: " + num)
            count +=1
    if count == 0:
       print("Sorry, the name is not found")
    

def SN(number):
    found = PB.get(number)
    if found != None:
        print("The Name is: " + found)
    else:
        print("Sorry, the number is not found")


option = -1 
while(option != '0'):
    option = input('Choose: 1-Search  2-Add/Update  3- Show')
   

    if option == '1':
        lookfor = input("Enter The Name/Number")
        if lookfor.isnumeric() and len(lookfor) == 10:
            SN(lookfor)
        elif lookfor.isalpha():
            SN(lookfor)
        else:
            print("This is invalid input")
    
    
    elif option == '2':
        new_name = input("Enter The Name: ")
        new_number = input("Enter The Number")
        if not new_name.isalpha():
            print("This is invalid name ")
        elif not new_number.isnumeric() or len(new_number) != 10:
            print("This is invalid number ")
        else:
            PB.update({new_number: new_name})
   

    elif option == '3':
        for names,nums in PB.items():
            print(names)
    
    
    else:
        print("Error")
```
